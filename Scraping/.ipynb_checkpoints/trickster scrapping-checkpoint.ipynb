# Imports

```python
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import urllib
import time
import shutil
import requests
```

# Save images to file

```python
def save_image_to_file(image, dirname, suffix):
    with open('{dirname}/trickster_{suffix}.png'.format(dirname=dirname, suffix=suffix), 'wb') as out_file:
        shutil.copyfileobj(image.raw, out_file)
```

# Scrapping

# Lvl 1 to 20

```python
url='https://wikimirror.lifeto.co/wiki.ggftw.com/trickster/Category_Level_001-020_Monster.html'
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(executable_path="../../chromedriver.exe", options=options)
driver.set_window_size(1120, 1000)

driver.get(url)
driver.implicitly_wait(10)

urls=[]
    
element=WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH, f"//ul//li//a[contains (@href,'.html')]")))
monsters=driver.find_elements_by_xpath(f"//ul//li//a[contains (@href,'.html')]")
   
nb=len(monsters)
    
for i in range(1,nb):

    href=monsters[i].get_attribute("href")
    if len(href.split('_'))<4:
        urls.append(href)
    
urls=urls[:23]

num=0

for url in urls:
    try:
        driver.get(url)
        img = WebDriverWait(driver, 15).until(
        EC.presence_of_element_located((By.XPATH, "//img[contains (@src,'.gif')]"))
        )

        img=driver.find_element_by_xpath("//img[contains (@src,'.gif')]")
        src = img.get_attribute('src')
        response = requests.get(src, stream=True)

        save_image_to_file(response,'trickster',num)
        num+=1
    
    except:
        print('Failed')
        continue
```
